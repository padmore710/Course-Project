#### 5. __*Data visualisation*__

```python
class RedWine:
    def __init__(self, dataset):
        self.dataset = pd.read_csv('red_wine.csv')
        self.summary = self.dataset.describe()

    def bar_plot(self):
        mean = self.summary.loc['mean']
        stan_dev = self.summary.loc['std']

        plt.figure(figsize=(8, 6))
        mean.plot(kind='bar', yerr=stan_dev, color='skyblue', alpha=0.7, capsize=5)
        plt.title('Mean with Standard Deviation')
        plt.ylabel('Values')
        plt.xlabel('Features')
        plt.xticks(rotation=45)
        plt.grid(axis='y')
        plt.tight_layout()
        plt.show()

    def box_plot(self):
        min_val = self.summary.loc['min']
        max_val = self.summary.loc['max']
        percentiles = self.summary.loc[['25%', '50%', '75%']]

        stats_data = pd.concat([min_val, max_val, percentiles])
        stats_data = stats_data.T

        plt.figure(figsize=(10, 6))
        stats_data.boxplot()
        plt.title('Min, Max, and Percentiles')
        plt.ylabel('Values')
        plt.xlabel('Statistics')
        plt.xticks(rotation=45)
        plt.grid(axis='y')
        plt.tight_layout()
        plt.show()

        
    def correlation_plot(self):
        correlation = self.dataset.corr()

        plt.figure(figsize=(10, 8))
        plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
        plt.colorbar()
        plt.title('Correlation Plot')
        plt.xticks(range(len(correlation)), correlation.columns, rotation=90)
        plt.yticks(range(len(correlation)), correlation.columns)
        plt.tight_layout()
        plt.show()
        
data = RedWine('red_wine.csv')
```

- ### Bar Chart
A bar chart to show the average values for different properties, with error bars representing the variability around these averages.

```python
data.bar_plot()
```

- ### Box Plot
This Box plot represents the range of values (min to max) and the spread of data using percentiles (25%, 50%, 75%) for each feature.

```python
data.box_plot()
```

- ### Correlation Plot
A correlation plot to visually capture the relationships between variables, where stronger correlations are indicated by warmer colours and values closer to 1. Where as cooler colours

```python
data.correlation_plot()
```
