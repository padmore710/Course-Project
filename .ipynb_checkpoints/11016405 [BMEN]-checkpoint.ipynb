```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
class RedWine:
    def __init__(self, dataset):
        self.dataset = pd.read_csv(dataset)

    def head_preview(self):
        return self.dataset.head(10)

    def tail_preview(self):
        return self.dataset.tail(20)

    def red_wine_info(self):
        return self.dataset.info()

    def red_wine_description(self):
        return self.dataset.describe()


data = RedWine('red_wine.csv')
```

```python
class RedWine:
    def __init__(self, dataset):
        self.dataset = pd.read_csv(dataset)

    def fa_fill(column, value):
        data.dataset[column] = data.dataset[column].fillna(value)

    def show_duplicates(self):
        return self.dataset.duplicated()

    def remove_duplicates(self):
        return self.dataset.drop_duplicates()    

data = RedWine('red_wine.csv')
```
